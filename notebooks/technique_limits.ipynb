````python
# Pattern to match headers (e.g., "4.1", "4.3.1", etc.)
header_pattern = re.compile(r"^4(?:\.\d+)+")

# Dictionary to hold buckets
buckets = dict()
current_header = None

buckets = {}
current_key = None

for line in filtered_data:
    if header_pattern.match(line):
        # Match header and set it as the current header
        current_header = line.split(" ", 1)[0]  # Extract header (e.g., "4.1")
        if current_header not in buckets:
            buckets[current_header] = []
        buckets[current_header].append(line)  # Add the header itself
    elif current_header:
        # Append line to the current header's bucket
        buckets[current_header].append(line)

# Display the buckets
new_buckets = dict()
for key, bucket in buckets.items():
    title = bucket[0].replace(f"{key} ", "").replace("Cottura al ", "").replace("Cottura ", "").replace("salto", "Saltare in Padella").replace("Grigliatura", "Grigliare")
    title = title.capitalize()
    if len(bucket) > 1:
        new_buckets[title] = bucket[1:]

for key, bucket in new_buckets.items():
    print(f"{key}:")
    for item in bucket:
        print(f"  {item}")
    print()

print(new_buckets.keys())


````

````python
from typing import List
from pydantic import BaseModel
from langchain_ollama import ChatOllama

class LicenceLLM(BaseModel):
    licence_type: str = "Licence Type"
    level: str = "Licence Numerical Level as string"

class LicenceListLLM(BaseModel):
    licence_list: List[LicenceLLM] = [LicenceLLM()]

llm = ChatOllama(
    model="llama3.2",
    temperature=0
)
structured_llm = llm.with_structured_output(schema=LicenceListLLM, method="json_mode")
````

````python
def possible_licenses():
    return ", ".join([f"`{licence.name}`" for licence in licences])

system_prompt = f"""
Your task is to extract from a TEXT a license type and level given POSSIBLE LICENSE TYPES.
If license is mentioned, the license level MUST always be present.
"""

user_prompt = """
POSSIBLE LICENSE TYPES:
{possible_licenses}

TEXT:
```
{text}
```
"""

format_prompt = f"""
Avoid the introductory and enclosing comments, return only JSON. Output must respect the following JSON format:
{LicenceListLLM().model_dump_json()}
"""
````

````python
from ingestor.technique_ingestor import TechniqueIngestor

class NewTechniqueSubCategory(BaseModel):
    name: str
    how_it_works: str
    pros: str
    cons: str
    limits: LicenceListLLM

ingestor = TechniqueIngestor()
pdf_path = "../Hackapizza Dataset/Misc/Manuale di Cucina.pdf"
techniques = ingestor.run(pdf_path=pdf_path)
result = list()
for macro_technique in techniques:
    for technique in macro_technique.techniques:
        technique_key = technique.category.capitalize()
        sub_categories_rules = new_buckets[technique_key]
        for rule,sub_category in zip(sub_categories_rules, technique.sub_categories):
            user_prompt_filled = user_prompt.format(
                possible_licenses=possible_licenses(),
                text=rule
            )
            print(user_prompt_filled)
            messages = [
                ("system", system_prompt),
                ("human", user_prompt_filled),
                ("human", format_prompt)
            ]

            try:
                output: LicenceListLLM = structured_llm.invoke(messages)
                result.append(NewTechniqueSubCategory(
                    name=sub_category.name,
                    how_it_works=sub_category.how_it_works,
                    pros=sub_category.pros,
                    cons=sub_category.cons,
                    limits=output
                ))
            except Exception as e:
                print(e)
````
